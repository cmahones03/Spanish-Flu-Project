```python
import csv
import os
import time
import requests
import pandas
```

```python
# Load URLs and names from CSV files
urls_csv = "SPFLU_The News Scimitar (Memphis) TEXT FILE URLS.csv"
names_csv = "The News Scimitar (Memphis TN 1918_19) issues containing keywords.csv"
output_dir = "SPFLU_The News Scimitar (Memphis) TEXT FILES"
os.makedirs(output_dir, exist_ok=True)

def get_data_from_csv(csv_path):
    """Reads a CSV file and returns a list of its first column values."""
    data = []
    with open(csv_path, newline='', encoding='utf-8') as file:
        reader = csv.reader(file)
        for row in reader:
            if row:  # Ensure the row is not empty
                data.append(row[0])  # Assuming relevant data is in the first column
    return data

def download_texts(urls, names):
    for idx, url in enumerate(urls):
        try:
            response = requests.get(url)
            response.raise_for_status()
            
            if idx < len(names):
                filename = names[idx].replace(" ", "_").replace(",", "").replace(".", "").replace("/", "_") + ".txt"
            else:
                filename = f"document_{idx+1}.txt"
            
            text_filepath = os.path.join(output_dir, filename)
            with open(text_filepath, 'w', encoding='utf-8') as text_file:
                text_file.write(response.text)
            
            print(f"Downloaded: {filename}")
            time.sleep(2)  # Delay to avoid overloading the server
        except Exception as e:
            print(f"Error processing {url}: {e}")

if __name__ == "__main__":
    urls = get_data_from_csv(urls_csv)
    names = get_data_from_csv(names_csv)
    download_texts(urls, names)
```
